```python
df = pd.read_csv("Data_Marketing_Customer_Analysis_Round3.csv")
```

```python
df_numeric = df.select_dtypes('int64')
df_numeric
```

I will drop the target which is "total_claim_amount"

```python
df_numeric1 = df_numeric.drop(['total_claim_amount'], axis=1)
```

```python
df_numeric1
```

```python
#Will reset index:
#df_numeric =df_numeric.reset_index(drop=True)
```

## x and y split

```python
X = df_numeric1
y = df_numeric['total_claim_amount']
```

```python
X
```

```python
y
```

### Next we do the train set split 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state = 42)
```

```python
X_train
```

## now we standarize
- scale train
scaler.fit(X_train)
scaler.transform(X_train) 
- scale test 
scaler.transform(X_test)

```python
from sklearn.preprocessing import PowerTransformer, StandardScaler, MinMaxScaler, OneHotEncoder, LabelEncoder
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
### Now we do the model generation
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
model = lm.fit(X_train_scaled,y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

### now we will test it 

```python
y_pred = model.predict(X_test_scaled)
```

```python
y_pred
```

```python
compare = pd.DataFrame({'y_test':list(y_test), 'y_pred':list(y_pred)})
```

```python
compare
```

## Model evaluation


```python
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse
```

```python
mse(y_test,y_pred)
```

```python
mae(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,3,figsize=(14,4))
ax[0].plot(y_pred, y_test, 'o')
ax[0].set_xlabel("y_test")
ax[0].set_ylabel("y_pred")
ax[0].set_title("Test Set -Predicted vs real")

# Get a histogram of the residuals ie: y - y_pred.  Homoscedasticity
# Does it resemble a normal distribution?
ax[1].hist(y_test - y_pred)
ax[1].set_xlabel("Test y-y_pred")
ax[1].set_title("Test Set Residual histogram")

ax[2].plot(y_pred,y_test - y_pred,"o")
ax[2].set_xlabel("predicted")
ax[2].set_ylabel("residuals")
ax[2].set_title("Residuals by Predicted")
ax[2].plot(y_pred,np.zeros(len(y_pred)),linestyle='dashed')
```

## Now we Rerun the model after adding the hot encoded categorical variables as well as other numeric categroical variables (e.g. number of open complaintes).

```python
df_categorical = df.select_dtypes('object')
df_categorical
```

```python
#first I will hot encode: 

columns_to_encode = ['region', 'response','employment_status','gender','location_code', 'marital_status','sales_channel','policy_type']
columns_hardencoded = pd.get_dummies(df_categorical, columns = columns_to_encode, drop_first = True)
columns_hardencoded



```

Now we start with the ordinal: 

```python
#Education
le = LabelEncoder()
le.fit(["high school or below", "college", "bachelor", "master", "doctor"])
df_categorical['education'] = le.transform(df_categorical['education'])
```

```python
#vehicle_class
le = LabelEncoder()
le.fit(["four-door car", "two-door car","sports car","luxury car", "suv", "luxury suv"])
df_categorical['vehicle_class'] = le.transform(df_categorical['vehicle_class'])
```

```python
#policy 
le = LabelEncoder()
le.fit(["personal l1", "personal l2", "personal l3", "special l1", "special l2","special l3","corporate l1", "corporate l2", "corporate l3"])
df_categorical['policy'] = le.transform(df_categorical['policy'])
```

```python
#Renew offer type 
le = LabelEncoder()
le.fit(["offer1", "offer2","offer3","offer4"])
df_categorical['renew_offer_type'] = le.transform(df_categorical['renew_offer_type'])
```

```python
#Vehicle size 
le = LabelEncoder()
le.fit(["small", "medsize","large"])
df_categorical['vehicle_size'] = le.transform(df_categorical['vehicle_size'])
```

```python
#coverage
le = LabelEncoder()
le.fit(["basic", "extended","premium"])
df_categorical['coverage'] = le.transform(df_categorical['coverage'])
```

```python
df_categorical
```

I will drop columns and then concat with the hard coded df

```python
ordinal_columns = df_categorical.drop(['region', 'response', 'effective_to_date', 'month', 'employment_status', 'marital_status','gender', 'location_code', 'policy_type', 'sales_channel'], axis=1)
```

```python
ordinal_columns
```

```python
ordinal_columns = ordinal_columns.reset_index(drop=True)
ordinal_columns
```

Now I will drop columns from my hardcoded df: 

```python
columns_hardencoded 
```

```python
columns_hardencoded = columns_hardencoded.drop(['coverage', 'education', 'effective_to_date', 'month', 'policy', 'renew_offer_type', 'vehicle_class', 'vehicle_size'], axis=1)
```

```python
columns_hardencoded
```

```python
columns_hardencoded = columns_hardencoded.reset_index(drop=True)
columns_hardencoded
```

Now I concat hardcoded columns and ordinal columns: 

```python
categorical_now_numeric = pd.concat([columns_hardencoded, ordinal_columns], axis=1)
categorical_now_numeric = categorical_now_numeric.reset_index(drop=True)
```

```python
categorical_now_numeric
```

```python
df_numeric1 = df_numeric1.reset_index(drop=True)
df_numeric1
```

```python
numeric_model_two = pd.concat([categorical_now_numeric, df_numeric1], axis=1)
numeric_model_two
```

```python
numeric_model_two = numeric_model_two.reset_index(drop=True)
numeric_model_two
```

# Now we run the model! 

```python
# X and y split 
X = numeric_model_two
y = df_numeric['total_claim_amount']
```

```python
# train and set split 
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state = 42)
```

```python
X_train
```

```python
#we standarize
from sklearn.preprocessing import PowerTransformer, StandardScaler, MinMaxScaler, OneHotEncoder, LabelEncoder
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#model regression
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
model = lm.fit(X_train_scaled,y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
#let's test it 
y_pred = model.predict(X_test_scaled)
```

```python
y_pred
```

```python
compare = pd.DataFrame({'y_test':list(y_test), 'y_pred':list(y_pred)})
compare
```
