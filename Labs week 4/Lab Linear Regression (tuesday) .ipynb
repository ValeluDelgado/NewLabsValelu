```python
df = pd.read_csv("Data_Marketing_Customer_Analysis_Round3.csv")
```

```python
df_numeric = df.select_dtypes('int64')
df_numeric
```

I will drop the target which is "total_claim_amount"

```python
df_numeric1 = df_numeric.drop(['total_claim_amount'], axis=1)
```

```python
df_numeric1
```

```python
#Will reset index:
#df_numeric =df_numeric.reset_index(drop=True)
```

## x and y split

```python
X = df_numeric1
y = df_numeric['total_claim_amount']
```

```python
X
```

```python
y
```

### Next we do the train set split 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state = 42)
```

```python
X_train
```

## now we standarize
- scale train
scaler.fit(X_train)
scaler.transform(X_train) 
- scale test 
scaler.transform(X_test)

```python
from sklearn.preprocessing import PowerTransformer, StandardScaler, MinMaxScaler, OneHotEncoder, LabelEncoder
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
### Now we do the model generation
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
model = lm.fit(X_train_scaled,y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

### now we will test it 

```python
y_pred = model.predict(X_test_scaled)
```

```python
y_pred
```

```python
compare = pd.DataFrame({'y_test':list(y_test), 'y_pred':list(y_pred)})
```

```python
compare
```

## Model evaluation


```python
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse
```

```python
mse(y_test,y_pred)
```

```python
mae(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,3,figsize=(14,4))
ax[0].plot(y_pred, y_test, 'o')
ax[0].set_xlabel("y_test")
ax[0].set_ylabel("y_pred")
ax[0].set_title("Test Set -Predicted vs real")

# Get a histogram of the residuals ie: y - y_pred.  Homoscedasticity
# Does it resemble a normal distribution?
ax[1].hist(y_test - y_pred)
ax[1].set_xlabel("Test y-y_pred")
ax[1].set_title("Test Set Residual histogram")

ax[2].plot(y_pred,y_test - y_pred,"o")
ax[2].set_xlabel("predicted")
ax[2].set_ylabel("residuals")
ax[2].set_title("Residuals by Predicted")
ax[2].plot(y_pred,np.zeros(len(y_pred)),linestyle='dashed')
```
