Show a plot of the total number of responses.
Response is an object and the answers are yes or no, so I will use a count. 

```python
responses = object_data['response'].value_counts()
responses
```

```python
sns.countplot(x="response", data=object_data)
```

Show a plot of the response by the sales channel.


```python

channel_response = data.groupby(['sales_channel', 'response']).size().reset_index(name='count')
channel_response


```


We then pivot the data to have the response categories ('yes' and 'no') as separate columns using the pivot function. 
Pivot --> to have response categories as columns

```python
pivot_response = channel_response.pivot(index='sales_channel', columns='response', values='count')
pivot_response
```

Here, we then plot the response by sales channel: 

```python

sns.set(style='whitegrid')
pivot_response.plot(kind='bar', stacked=True)
plt.xlabel('Sales Channel')
plt.ylabel('Count')
plt.title('Response by Sales Channel')
plt.legend(title='Response')
plt.show()
```

```python
bins = [0, 200, 400, 600, 800, 1000, 1200, 1400, float('inf')]  # Specify the bin edges
labels = ['<200', '200-400', '400-600','600-800', '800-1000', '1000-1200', '1200-1400', '>1400']
data['claim_range'] = pd.cut(data['total_claim_amount'], bins=bins, labels=labels)
```
