Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

df1 = total_claim_amount > than $1,000 and response = 'yes'


```python
df1 = data.loc[(data['total claim amount']>1000) & (data['response']=='Yes')]
df1
```

2. Analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign.

groupby ( policy type, then gender)(total claim amount).mean 

average of total claim amount by policy type and gender and yes to campaign 

```python
df1.groupby(["policy type","gender","response"], as_index= False)['total claim amount'].mean()
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

total customers in each state and THEN filter of states > 500

```python
data.state
```

```python
data.groupby('state').filter(lambda x: x['state'].count()>500)
```

```python
states = data.groupby(['state'])['state'].count().rename('count').reset_index()
states
```

```python
states.loc[states['count']>500]
```

4. Find the median customer lifetime value by education level and gender.

lifetime value median by education y gender 

```python
data.groupby(["education","gender"], as_index= False)['customer lifetime value'].median()
```
